```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns
import os
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
df = pd.read_csv('md_df.csv')
```

```python
from sklearn import preprocessing


scaler = preprocessing.StandardScaler()
x_scale = preprocessing.scale(df)
```

```python
inputs = x_scale.shape[1]

autoencoder = keras.Sequential(
    [
       layers.Dense(int(inputs/2), activation = 'relu', input_shape = (inputs,), name = 'input_layer'),
        layers.Dense(int(inputs/4), activation = 'relu', name = 'hidden1'),
        layers.Dense(2, activation = 'relu', name = 'code'),
        layers.Dense(int(inputs/4), activation = 'relu', name = 'hidden2'),
        layers.Dense(int(inputs/2), activation = 'relu', name = 'hidden3'),
        layers.Dense(inputs, activation = 'relu', name = 'output')
    
])
autoencoder.summary()
autoencoder.compile(optimizer = 'adam', loss = 'mse')
history = autoencoder.fit(x_scale, x_scale, epochs = 50, verbose = 1)
```
